```python
import pandas as pd
```

```python
df =  pd.read_csv("FMCG Raw Sales Data.csv")
```

```python
df
```

```python
df["month"] = df["Month"].apply(lambda x : x[x.index("-")+1:-3])
```

```python
df
```

```python
def clean_amount(amount):
    
    return int(amount.replace(",",""))
    
```

```python
df["Qty"] = df["Qty"].apply(clean_amount)
df["Sales"] = df["Sales"].apply(clean_amount)
```

```python
df.groupby("Category")["Qty"].sum()
```

```python
df.groupby("Sub-Channels")["Sales"].sum()
```

```python
df[df["Sales"] == df["Sales"].max()]["month"]
```

```python
def clean(prdct):
    for i in prdct:
        if i.isdigit()==True:
           return prdct[:prdct.index(i)].strip()
```

```python
df["products"] =  df["Product Name"].apply(clean)
```

```python
df
```

```python
pd.DataFrame(df["products"].unique())
```

```python
df[df["Sales"]==df["Sales"].max()]["products"]
```

```python
with pd.ExcelWriter("project3.xlsx",engine = "openpyxl") as writer :
    df.to_excel(writer , sheet_name = "month name" , index = False  )
    (df.groupby("month")["Qty"].sum()).to_excel(writer,sheet_name = "monthwise sum of Qty")
    df.groupby("Category")["Qty"].sum().to_excel(writer,sheet_name = "category wise sum of Qty")
    df[df["Sales"] == df["Sales"].max()]["month"].to_excel(writer,sheet_name = " highest sales month")
    df[df["Sales"]==df["Sales"].max()]["products"].to_excel(writer,sheet_name = "highest saling product")
    df.to_excel(writer,sheet_name= "clean product name",index = False)
    pd.DataFrame(df["products"].unique()).to_excel(writer, sheet_name ="unique product name")
    df.groupby("Sub-Channels")["Sales"].sum().to_excel(writer,sheet_name="sub-channel wise total sales")
```
