```python
import pandas as pd
```

```python
df = pd.read_csv("orders.csv")
```

```python
df
```

```python
df.info()
```

```python
df["Exequted Qty"] = df['Qty.'].apply( lambda x : int (x.split("/")[0]))
```

```python
df['Total price'] = df["Exequted Qty"]* df["Avg. price"]
```

```python
df
```

```python
df = df[df["Product"]=="MIS"]
df = df[df["Status"]=='COMPLETE']
```

```python
df.rename(columns = {"Total price":"Turnover"}, inplace = True)
```

```python
df
```

```python
df["ETC"] = df.apply(lambda row : row["Turnover"]*0.0000297,axis=1).round(2)
```

```python
df["Brokrage"] = df.apply(lambda x : x["Turnover"]*0.0003, axis=1).round(2)
```

```python
df["STT/CTT"] = df.apply(lambda x : x["Turnover"]*0.00025 if x["Type"]=="SELL" else 0  , axis=1).round(2)
```

```python
df["SEBI"] = df.apply(lambda x : x["Turnover"]*0.000001, axis=1).round(2)
```

```python
df["GST"] = df.apply(lambda x : x["Brokrage"]*0.18,axis=1).round(2)
```

```python
df["STAMP_charge"] =df.apply(lambda x : x["Turnover"]*0.00003 if x["Type"]=="BUY" else 0, axis=1).round(2)
```

```python
df["Total charges"] = df.apply(lambda x : x["ETC"]+x["Brokrage"]+x["STT/CTT"]+x["SEBI"]+x["GST"]+x["STAMP_charge"], axis=1).round(2)
```

```python
grouped_df = df.groupby(["Instrument", "Type"])[
    ["Exequted Qty", "Avg. price", "Brokrage", "STT/CTT", "SEBI", "GST",
     "ETC", "STAMP_charge", "Total charges"]
].sum()


```

```python
grouped_df
```

```python
df["Buy"] = df.apply(lambda df : df["Turnover"] if df["Type"]=="BUY" else 0 , axis=1)
df["Sell"] = df.apply(lambda df : df["Turnover"] if df["Type"]=="SELL" else 0 , axis=1)
```

```python
df
```

```python
df3 = df.groupby("Instrument")[
["Exequted Qty","Buy","Sell","Total charges"]
].sum()
```

```python
df3["gross pnl"] = df3["Sell"] - df3["Buy"]
```

```python
df3["Net pln"] = df3["gross pnl"] - df3["Total charges"]
```

```python
df3["% Charges on Gross PnL"] =((df3["Total charges"]/df3["gross pnl"])*100).round(2)
```
